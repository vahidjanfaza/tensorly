```python
import tensorly as tl
from tensorly.decomposition import tucker, parafac, non_negative_tucker
import torch
from tensorly import tucker_to_tensor
import numpy as np
```

```python
in_rand = np.random.rand(512,512,3,3)
input1 = tl.tensor(in_rand)
```

```python
core, factors = non_negative_tucker(input1, rank=[512,512,3,3], n_iter_max=40, init='svd')
# print(core)
# print(factors)
```

```python
core_torch = np.array(core)
print(core.shape)
factors_torch = []
factors_torch.append(np.array(factors[0]))
factors_torch.append(np.array(factors[1]))
factors_torch.append(np.array(factors[2]))
factors_torch.append(np.array(factors[3]))
[f.shape for f in factors]
```

```python
ORG_Mat = np.einsum('ijkl,mi->mjkl', core_torch, factors_torch[0])
print(ORG_Mat.shape)
ORG_Mat = np.einsum('ijkl,mj->imkl', ORG_Mat, factors_torch[1])
print(ORG_Mat.shape)
ORG_Mat = np.einsum('ijkl,mk->ijml', ORG_Mat, factors_torch[2])
print(ORG_Mat.shape)
ORG_Mat = np.einsum('ijkl,ml->ijkm', ORG_Mat, factors_torch[2])
print(ORG_Mat.shape)
```

```python
np.max(ORG_Mat-input1)
```

```python
print(ORG_Mat-input1)
```
